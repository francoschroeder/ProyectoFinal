```python
import datetime as dt
from datetime import datetime
import pandas as pd
```

```python
dateparse = lambda x: datetime.strptime(x, '%d.%m.%y %H:%M')

data = pd.read_csv("regina.csv", parse_dates=['Date Time'], date_parser=dateparse, index_col=0)
```

```python
file = open("brokenDates.txt", "x")

for index, row in data.iterrows():
    if row['Temp Out'] == '---':
        print(index)
        file.write((index).strftime("%d/%m/%y, %H:%M:%S") + "\n")

file.close()
```
